```python
from bs4 import BeautifulSoup as soup  # HTML data structure
from urllib.request import urlopen as uReq  # Web client

```

```python
# URl to web scrap from.
# in this example we web scrap graphics cards from Newegg.com
#page_url = "http://www.newegg.com/Product/ProductList.aspx?Submit=ENE&N=-1&IsNodeId=1&Description=GTX&bop=And&Page=1&PageSize=36&order=BESTMATCH"
page_url = "https://www.newegg.com/Video-Cards-Video-Devices/Category/ID-38?Tpk=graphics%20cards"
#page_url = "https://nqrural.com.au/chemicals.html"
#page_url = "https://www.newegg.com/Workstation-Graphics-Cards/SubCategory/ID-449"


# opens the connection and downloads html page from url
uClient = uReq(page_url)

# parses html into a soup data structure to traverse html
# as if it were a json data type.
page_soup = soup(uClient.read(), "html.parser") #html parsing
uClient.close()

```

```python
filename = "Videocards.csv"
f= open(filename, "w")

headers = "Brand_name, Product_name, shipping_price\n"

f.write(headers)

for container1 in containers:
    Brand_name = container1.findAll("a", {"class": "item-brand"})[0].img["title"]
    
    Product_name = container1.a.img["title"].title()
    
    shipping = container1.findAll("li", {"class": "price-ship"})
    shipping_price = shipping[0].text.strip()
    
    print("Brand_name:" + Brand_name)
    print("Product_name:" + Product_name)
    print("shipping_price:" + shipping_price)
    
    f.write(Brand_name + "," +Product_name.replace(",", "|") + "," + shipping_price +"\n")
    
f.close()
```
